```python
import numpy as np
```

```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
a = np.array([0,1,2,3,4,5])
b = np.array([0,1706,650,1092,1230,706])
plt.figure(figsize=(10,5))
plt.title ("2016 - 2018 Accident per Province in CALABARZON")
plt.xlabel ("Provinces")
plt.ylabel ("Number of Accidents")
plt.plot(a,b)
plt.xticks(range(0,len(a)),('','Cavite','Laguna','Batangas','Rizal','Quezon'))
plt.savefig("Accident per Province in Calabarzon")
plt.show()
```

```python
province = ['Cavite','Laguna','Batangas','Rizal','Quezon']
numofdeaths = np.array([11,0,6,6,1])
plt.figure(figsize=(10,5))
plt.title ("2016 - 2018 Deaths per Province in CALABARZON")
plt.xlabel ("Provinces")
plt.ylabel ("Number of Deaths")
plt.bar(range(0,5),numofdeaths)
plt.xticks(range(0,5),province)
plt.savefig("Deaths per Province in CALABARZON")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title ("Accidents Per Day and Time")
plt.plot(raw_data.Category, raw_data.Morning, '--', label = "Morning")
plt.plot(raw_data.Category, raw_data.Afternoon, '--',label = "Afternoon")
plt.plot(raw_data.Category, raw_data.Early_Night, label = "Early Night")
plt.plot(raw_data.Category, raw_data.Late_Night, label = "Late Night")
plt.legend(loc="upper center")
plt.xlabel ("Days")
plt.ylabel ("Number of Accidents")
plt.savefig("Accident per Day and Time in CALABARZON")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title ("Accidents Per Gender and Age Group")
plt.plot(raw_data1.Category, raw_data1.Female, ".-" ,color = "Red",label = "Female")
plt.plot(raw_data1.Category,raw_data1.Male, ".-", color = "Blue",label = " Male")
plt.xlabel ("Age Group")
plt.ylabel ("Number of Accidents")
plt.legend()
plt.savefig("Accident per Gender and Age in CALABARZON")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title ("Number of Deaths Per Age")
plt.bar(raw_data1.Category,raw_data1.Death_M,color = "Blue",label = " Male")
plt.bar(raw_data1.Category, raw_data1.Death_F,color = "Red",label = "Female")
plt.xlabel ("Age Group")
plt.ylabel ("Number of Deaths")
plt.savefig("Number of Deaths per Age in CALABARZON")
plt.legend()

plt.show()
```
